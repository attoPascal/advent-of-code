```python
import numpy as np
from aocd import data
```

```python
grid = np.array([list(line) for line in data.splitlines()])
```

```python
(x, *_), (y, *_) = np.where(grid == 'S')

start = (x, y)
loop = [start]
previous = start

# hardcoded yolo
grid[*start] = '|' 
pipe = (x+1, y)

while pipe != start:
    loop.append(pipe)
    x, y = pipe
    n, s, e, w = (x-1, y), (x+1, y), (x, y+1), (x, y-1)
    connect = {
        '|': (n, s),
        '-': (e, w),
        'L': (n, e),
        'J': (n, w),
        '7': (s, w),
        'F': (s, e)
    }[grid[x, y]]
    pipe, previous = connect[connect[0] == previous], pipe

print("Part 1:", len(loop) // 2)
```

```python
count_inside = 0
is_inside = False
pipes = set(loop)
prev_bend = ''

for (x, y), tile in np.ndenumerate(grid):
    if (x, y) in pipes:
        if tile == '|':
            is_inside = not is_inside
        elif tile in 'LF':
            prev_bend = tile
        elif tile in 'J7':
            bend = prev_bend + tile
            if bend in ['FJ', 'L7']: # bends like ┌─┘ or └─┐
                is_inside = not is_inside
    else:
        count_inside += is_inside

print("Part 2:", count_inside)
```

```python
pretty = np.full_like(grid, ' ')
for (x, y) in loop:
    pretty[x, y] = {
        '-': '─',
        '|': '│',
        'F': '┌',
        '7': '┐',
        'L': '└',
        'J': '┘'
    }[grid[x, y]]

for line in pretty:
    print(''.join(line))
```
