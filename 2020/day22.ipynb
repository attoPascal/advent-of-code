```python
from itertools import count
```

```python
game_count = 1
def play_game(player1, player2, verbose=False):
    global game_count
    game = game_count
    game_count += 1

    if verbose:
        print("Game", game, "starts")
    configurations = []

    for round in count(1):
        if verbose:
            print("Game", game, "Round", round)
            print("Player 1:", player1)
            print("Player 2:", player2)

        if player1 in configurations:
            if verbose:
                print("Player 1 wins by repetition")
                print(player1)
                print(configurations)
            return (1, player1)
        
        configurations.append(player1.copy())
        card1 = player1.pop(0)
        card2 = player2.pop(0)
        if verbose:
            print("Cards drawn:", card1, card2)

        if len(player1) < card1 or len(player2) < card2:
            if card1 > card2:
                player1.extend([card1, card2])
                if verbose:
                    print("Player 1 wins round", round)
            else:
                player2.extend([card2, card1])
                if verbose:
                    print("Player 2 wins round", round)
        else:
            player1_sub = player1.copy()[:card1]
            player2_sub = player2.copy()[:card2]
            if verbose:
                print("Sub-game...")
            win, deck = play_game(player1_sub, player2_sub)
            if win == 1:
                player1.extend([card1, card2])
                if verbose:
                    print("Player 1 wins round", round)
            else:
                player2.extend([card2, card1])
                if verbose:
                    print("Player 2 wins round", round)
        
        if player1 == []:
            if verbose:
                print("Player 2 wins game", game)
            return (2, player2)
        elif player2 == []:
            if verbose:
                print("Player 1 wins game", game)
            return (1, player1)
        
player1 = [8, 19, 46, 11, 36, 10, 35, 9, 24, 22, 50, 1, 34, 7, 18, 28, 3, 38, 43, 2, 6, 42, 23, 12, 20]
player2 = [39, 27, 44, 29, 5, 48, 30, 32, 15, 31, 14, 21, 49, 17, 45, 47, 16, 26, 33, 25, 13, 41, 4, 40, 37]
win, deck = play_game(player1, player2, verbose=True)
```

```python
sum(c*p for c, p in zip(deck, range(50, 0, -1)))
```
