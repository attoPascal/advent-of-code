```python
class Ship1():
    def __init__(self):
        self.pos_h = 0
        self.pos_v = 0
        self.deg = 0
    
    @property
    def pos(self):
        return abs(self.pos_h) + abs(self.pos_v)
    
    @property
    def dir(self):
        return {
            0: 'E',
            90: 'N',
            180: 'W',
            270: 'S'
        }[self.deg  % 360]

    def move(self, dir, dist):
        if dir == 'F':
            dir = self.dir
        
        move_h, move_v = {
            'N': (-dist, 0),
            'E': (0, dist),
            'S': (dist, 0),
            'W': (0, -dist)
        }[dir]

        self.pos_h += move_h
        self.pos_v += move_v
    
    def turn(self, dir, deg):
        if dir == 'L':
            self.deg += deg
        else:
            self.deg -= deg
    
    def parse(self, instruction):
        letter = instruction[0]
        amount = int(instruction[1:])

        if letter in ['L', 'R']:
            self.turn(letter, amount)
        else:
            self.move(letter, amount)
```

```python
with open('input12.txt') as f:
    lines = f.read().splitlines()
```

```python
ship = Ship1()
for line in lines:
    ship.parse(line)
ship.pos
```

```python
class Ship2():
    def __init__(self):
        self.position  =  0 + 0j
        self.direction = 10 - 1j
    
    @property
    def manhattan_distance(self):
        return round(abs(self.position.real) + abs(self.position.imag))
    
    def turn_waypoint(self, dir, deg):
        turns = deg // 90 if dir == 'R' else -deg // 90
        self.direction *= 1j ** turns

    def move_ship(self, dist):
        self.position += dist * self.direction

    def move_waypoint(self, dir, dist):
        self.direction += {
            'N': -dist*1j,
            'E': dist,
            'S': dist*1j,
            'W': -dist
        }[dir]
    
    def parse(self, instruction):
        letter = instruction[0]
        amount = int(instruction[1:])

        if letter == 'F':
            self.move_ship(amount)
        elif letter in ['L', 'R']:
            self.turn_waypoint(letter, amount)
        else:
            self.move_waypoint(letter, amount)
```

```python
ship = Ship2()
for line in lines:
    ship.parse(line)
ship.manhattan_distance
```

```python
ship = Ship2()
for line in lines:
    ship.parse(line)
    print(line)
    h = ship.position.real
    v = ship.position.imag
    print(f"POS: {abs(h):>3} {'E' if h > 0 else 'W'}, {abs(v):>3} {'S' if v > 0 else 'N'}")

    h = ship.direction.real
    v = ship.direction.imag
    print(f"DIR: {abs(h):>3} {'E' if h > 0 else 'W'}, {abs(v):>3} {'S' if v > 0 else 'N'}")
    print()
```
