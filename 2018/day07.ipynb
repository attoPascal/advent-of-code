```python
import networkx as nx
import re
import aocd
from itertools import count
```

```python
lines = aocd.get_data(day=7, year=2018).splitlines()
G = nx.DiGraph()

for line in lines:
    m = re.match(r'.* ([A-Z]) .* ([A-Z])', line)
    u = m.group(1)
    v = m.group(2)
    G.add_edge(u, v)

nx.draw(G, with_labels=True)
```

```python
order = ''

while len(G) > 0:
    candidates = sorted([node for node, in_degree in G.in_degree() if in_degree == 0])
    next_step = candidates[0]
    order += next_step
    G.remove_node(next_step)

order
```
